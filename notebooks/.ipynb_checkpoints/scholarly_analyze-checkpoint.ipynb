# Collaboration Data

```python
with open('output_scholarly.pkl', 'rb') as input:
    list_scholarly = pickle.load(input)
```

### Create list of collaboration pairs between faculty based on shared publications

```python
dict_pubs = {}

for i in range(len(list_scholarly)):
    if list_scholarly[i] != 'None':
        author = list_scholarly[i]['scholar_id']
        list_pubs = list_scholarly[i]['publications']
        num_pubs = len(list_pubs)
        
        for i in range(num_pubs):
            pub_title = list_pubs[i]['bib']['title']
            
            if pub_title in dict_pubs and author not in dict_pubs[pub_title]:
                dict_pubs[pub_title].append(author)
            else:
                dict_pubs[pub_title] = [author]
```

```python
dict_pubs_filter1 = {key:value for key,value in dict_pubs.items() if len(value) > 1}
print(len(dict_pubs_filter1))
print(list(dict_pubs_filter1.items())[:20])
```

```python
authors_2 = Counter()

for authors in list(dict_pubs_filter1.values()):
    authors.sort()
    for comb in combinations(authors, 2):
        authors_2[comb] += 1

authors_2.most_common()[:20]
```

```python
list_collabPairs = [key for key, _ in authors_2.most_common()]
list_collabPairs[:20]
```

### Create dictionary of collaborations for each faculty

```python
dict_collab = {}

for pair in list_collabPairs:
    faculty1 = pair[0]
    faculty2 = pair[1]
    
    if faculty1 in dict_collab and faculty2 not in dict_collab[faculty1]:
        dict_collab[faculty1].append(faculty2)
    else:
        dict_collab[faculty1] = [faculty2]
        
    if faculty2 in dict_collab and faculty1 not in dict_collab[faculty2]:
        dict_collab[faculty2].append(faculty1)
    else:
        dict_collab[faculty2] = [faculty1]

list(dict_collab.items())[:20]
```

### Replace scholar ID with name and create list of collaboration for df

```python
map_facultyID = dict(zip(df_listFaculty['Scholar ID'], df_listFaculty['Title']))
map_facultyID
```

```python
list_collab = []

for scholar in list(df_listFaculty['Scholar ID']):
    try: 
        if scholar != "None":
            collabs = [map_facultyID[x] for x in dict_collab[scholar]]
            list_collab.append(collabs)
        else:
            list_collab.append([])
    except:
        list_collab.append([])

list_collab
```

```python
df_listFaculty['Collaboration'] = list_collab
df_listFaculty
```
